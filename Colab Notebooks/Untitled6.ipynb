```python
operacion = input("Ingrese la operación que desea realizar (+, -, *, /) o 'S' para salir: ")

while operacion != 'S':
    a = float(input('Ingrese el primer número: '))
    b = float(input('Ingrese el segundo número: '))

    if operacion == '+':
        resultado = a + b
    elif operacion == '-':
        resultado = a - b
    elif operacion == '*':
        resultado = a * b
    elif operacion == '/':
        resultado = a / b
    else:
        print('Operación no válida')

    print('El resultado es:', resultado)

    operacion = input("Ingrese la operación que desea realizar (+, -, *, /) o 'S' para salir: ")

```
