```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

features = pd.read_csv('/content/drive/MyDrive/BIG DATA CURSO/SEMANA 4/Features.csv')
streams = pd.read_csv('/content/drive/MyDrive/BIG DATA CURSO/SEMANA 4/Streams.csv')

features = features.rename(columns={'name': 'Song'})
final = pd.merge(streams,features,how = 'left', on = 'Song') #lef join de las tablas 'features' y 'streams'
final['duration'].plot.hist()

filtro = (final['duration'] >= 3) & (final['duration'] <= 6)

resultado = final['duration'][filtro]

print(resultado)
resultado.plot.hist()
```
